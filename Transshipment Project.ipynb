# IMPORTING GUROBI

```python
from gurobipy import *

milp_model = Model("milp")
```

# OPTIMIZING MODEL

```python
a={
    
}
b={
    
}
for i in range(3000):
    # GIVING PARAMETER VALUES
    import numpy as np
    from numpy import random
    b1 = random.randint(1,20)
    b2 = random.randint(1,30)
    b3 = random.randint(35,50)
    b4 = random.randint(45,60)

    d1 = random.randint(b1)
    d2 = random.randint(b2)
    d3 = random.randint(b3+1,52)
    d4 = random.randint(b4+1,62)

    C13 = random.randint(1,5)
    C14 = random.randint(1,5)
    C23 = random.randint(1,5)
    C24 = random.randint(1,5)
      
    while (b1+b2+b3+b4) < (d1+d2+d3+d4):
        b1 = random.randint(1,20)
        b2 = random.randint(1,30)
        b3 = random.randint(35,50)
        b4 = random.randint(45,60)
        d1 = random.randint(b1)
        d2 = random.randint(b2)
        d3 = random.randint(b3+1,52)
        d4 = random.randint(b4+1,62)       
        if (b1+b2+b3+b4) >= (d1+d2+d3+d4):
            break
            
    # DECISION VARIABLE
    x13  = milp_model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x13")
    x14 = milp_model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x14")
    x23 = milp_model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x23")
    x24 = milp_model.addVar(vtype=GRB.CONTINUOUS, lb=0, name="x24")
    
    # OBJECTIVE FUNCTION

    obj_fn = ((C13*x13)+(C14*x14)+(C23*x23)+(C24*x24))   
    milp_model.setObjective(obj_fn, GRB.MINIMIZE)
    
    # CONSTRAINTS
    c1 = milp_model.addConstr(x13+x14<=b1-d1, "c1")
    c2 = milp_model.addConstr(x23+x24<=b2-d2, "c2")
    c3 = milp_model.addConstr(x13+x23>=d3-b3, "c3")
    c4 = milp_model.addConstr(x14+x24>=d4-b4, "c4")
    
    milp_model.optimize()
    
    a[i]=[b1,b2,b3,b4,d1,d2,d3,d4,C13,C14,C23,C24]
    
    b[i]=[x13.x,x14.x,x23.x,x24.x]
```

# CREATING DATAFRAME

```python
import pandas as pd
df = pd.DataFrame(a, index=['b1','b2','b3','b4','d1','d2','d3','d4','C13','C14','C23','C24'])
df1=df.T
df2=pd.DataFrame(b, index=['x13','x14','x23','x24'])
df3=df2.T
```

```python
data = pd.concat([df1, df3], axis=1, join='inner')
```
